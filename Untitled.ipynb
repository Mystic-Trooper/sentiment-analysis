```python
data=data.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))
```

```python
data['tweet_clean'] = data['text'].replace(r'http\S+', '', regex=True).replace(r'www\S+', '', regex=True)
data.sample(5)
```

```python
import string
string.punctuation

#punctuation marks

def remove_punct(text):
    text_nopunct = "".join([char for char in str(text) if char not in string.punctuation])
    return text_nopunct

data['tweet_clean'] = data['tweet_clean'].apply(lambda x: remove_punct(x))


data.head()
```

```python
data['tweet_clean'] = data['tweet_clean'].str.lower()
data.head()
```

```python
data['tweet_clean'] = data['tweet_clean'].str.replace('\d+', '')
```

```python
def remove_stopwords(txt):
            stop_words = set(stopwords.words('english'))
            stop_words.update(("mon","tue","wed","thu","fri","sat","sun","sunday","monday","tuesday","thursday","friday","saturday","sunday","thurs","thur","tues"))
            stop_words.update(("january","february","march","april","may","june","july","august",
              "september","october","november","december","jan","feb","mar","apr",
              "may","jun","jul","aug","sep","oct","nov","dec", "twitter", "thanking","thanks"))
    
            word_tokens = nltk.word_tokenize(txt) 
            filtered_sentence = ' '.join([w for w in word_tokens if not w in stop_words])
            return filtered_sentence
data['tweet_clean'] = data['tweet_clean'].apply(lambda x: remove_stopwords(x))

data.head()
```

```python
bagofwords=[]
for i in range(0,5000):
  words=corpus[i].split()
  bagofwords.append(words)
```

```python
print(bagofwords)
```

```python
import itertools
import collections
all_words=list(itertools.chain(*bagofwords))
print(all_words[0:10])
print()
print('most commanly occuring words and frequency\n')
words_freq=collections.Counter(all_words)
words_freq.most_common(20)
```

```python
freq_word_df=pd.DataFrame(words_freq.most_common(20),columns=['words','Freq'])
freq_word_df.head()
```

```python
fig,ax=plt.subplots(figsize=(15,10))
freq_word_df.sort_values(by='Freq').plot.barh(x='words',y='Freq',ax=ax,color='red')
plt.show();
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer=TfidfVectorizer (max_features=2500,min_df=7,max_df=0.8,stop_words=stopwords.words('english'))
processed_features=vectorizer.fit_transform(corpus).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(processed_features,data.sentiment,test_size=0.2,random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
text_classifier=RandomForestClassifier(n_estimators=200,random_state=0)
text_classifier.fit(X_train,y_train)
```

```python
predictions=text_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
print(accuracy_score(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
import pylab as plt
import matplotlib.pyplot as plt
import seaborn as sns
labels = [-1,0,1]
cmx = confusion_matrix(y_test, predictions)
print(cmx)
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cmx)
plt.title("Confusion Matrix of Decision Tree")
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
